```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

Hyperparameter search

```python
data = []

with open("./data/44BasicEnvPpo_beta_0.0001_RollerAgent.json", mode="r") as json_file:
    data.append(json.load(json_file))
    data[0].append(0.0001)
    
with open("./data/45BasicEnvPpo_beta_0.0005_RollerAgent.json", mode="r") as json_file:
    data.append(json.load(json_file))
    data[1].append(0.0005)

with open("./data/46BasicEnvPpo_beta_0.001_RollerAgent.json", mode="r") as json_file:
    data.append(json.load(json_file))
    data[2].append(0.001)

with open("./data/47BasicEnvPpo_beta_0.005_RollerAgent.json", mode="r") as json_file:
    data.append(json.load(json_file))
    data[3].append(0.005)

with open("./data/48BasicEnvPpo_beta_0.01_RollerAgent.json", mode="r") as json_file:
    data.append(json.load(json_file))
    data[4].append(0.01)

data[0]
    
```

```python
for entry in data:
    X = [elem[1] for elem in entry[0:-1]]
    Y = [elem[2] for elem in entry[0:-1]]
    plt.plot(X, Y, label=f"beta = {entry[-1]}")

plt.ylim(0, 1)
plt.legend()
plt.xlabel("Step")
plt.ylabel("Reward")
plt.title("Variable beta values, Basic Env PPO")
plt.grid()

plt.savefig("beta.pdf")
```
